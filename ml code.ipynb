```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from gensim.models import Word2Vec

def find_top_matches(concatenated_data, user_input):
    """
    Find the top matches for user input skills in the concatenated_data DataFrame and print the results.

    Parameters:
    concatenated_data (pd.DataFrame): DataFrame containing 'transformed_data' and 'designation' columns.
    user_input (str): Comma-separated string of user skills.
    """
    # Clone only the transformed_data column
    cloned_df = concatenated_data[['transformed_data']].copy()

    # Add user input to the last row of transformed_data column
    cloned_df.loc[len(cloned_df)] = [user_input]

    # Convert the transformed_data column to list of lists
    sentences = [row.split(",") for row in cloned_df['transformed_data']]

    # Train Word2Vec model
    model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)

    # Get Word2Vec embeddings for user input
    user_input_vector = [model.wv[word] for word in user_input.split(",") if word in model.wv]
    user_input_vector = sum(user_input_vector) / len(user_input_vector)

    # Calculate cosine similarity of the user input vector with all other vectors
    cosine_similarities = []
    for vector in sentences[:-1]:
        cosine_similarities.append(cosine_similarity([user_input_vector], [model.wv[word] for word in vector if word in model.wv])[0][0])

    # Get indices of the top matches
    top_matches_indices = sorted(range(len(cosine_similarities)), key=lambda i: cosine_similarities[i], reverse=True)[:20]

    # Get designations of the top matches
    top_matches = concatenated_data.iloc[top_matches_indices][['designation']].reset_index(drop=False)

    # Count the unique designations
    unique_designations = top_matches['designation'].value_counts()

    # Print the top matches
    print("Top Matches:")
    print(top_matches)

    # Print the unique designations
    print("\nUnique Designations:")
    print(unique_designations)

# Example usage:
# find_top_matches(concatenated_data, "python,azure")



```
